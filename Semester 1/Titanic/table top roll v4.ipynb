What is the chance that you will draw a dual land or a basic land on the first draw? There are 7 cards in a hand, and the deck has 100 cards.

You have a 1/7 chance to draw either a basic land or a dual land at the start on your first hand. There are 7 cards in a hand, and one of them should be a land?

```python
import random
import pandas as pd
import seaborn as sns

# Read CSV into a DataFrame
df = pd.read_csv('eldrazi.csv')

# Run 10 times
for i in range(100):
    drawn_items = df.sample(n=7)
    print(f"Draw {i+1}:")
    print(drawn_items)
    print("-" * 80)
```
