What base evolution pokemon has the highest speed stat in generation one? This data set was created by a group of people on Kagle. It only includes the stats of the pokemon in generation one.

```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv("evo.csv")

# The speed stat seems to be under "Base Speed"
if "Base Speed" in df.columns:
    speed = df["Base Speed"].dropna()
else:
    raise KeyError("Couldn't find a column for speed.")

# Calculate measures of central tendency and spread
mean_speed = speed.mean()
median_speed = speed.median()
mode_speed = speed.mode().iloc[0]
std_speed = speed.std()
var_speed = speed.var()

# Plot histogram with matplotlib (not seaborn as per user request: pandas + matplotlib only)
plt.figure(figsize=(10, 6))
plt.hist(speed, bins=20, edgecolor="black", alpha=0.7)
plt.axvline(mean_speed, color="red", linestyle="--", label=f"Mean = {mean_speed:.2f}")
plt.axvline(median_speed, color="green", linestyle="--", label=f"Median = {median_speed:.2f}")
plt.axvline(mode_speed, color="blue", linestyle="--", label=f"Mode = {mode_speed:.2f}")

plt.title("Histogram of Base Speed")
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.legend()
plt.show()

(mean_speed, median_speed, mode_speed, std_speed, var_speed)
```

```python
# Assistant
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("evo.csv")

# The speed stat seems to be under "Base Speed"
if "Base Speed" in df.columns:
    speed = df["Base Speed"].dropna()
else:
    raise KeyError("Couldn't find a column for speed.")

# Calculate measures of central tendency
mean_speed = speed.mean()
median_speed = speed.median()
mode_speed = speed.mode().iloc[0]

# Calculate measures of spread
std_speed = speed.std()
var_speed = speed.var()
range_speed = speed.max() - speed.min()
iqr_speed = speed.quantile(0.75) - speed.quantile(0.25)
# Replace mad() with manual calculation of mean absolute deviation
mad_speed = (speed - mean_speed).abs().mean()  # Manually calculate MAD

# Plot histogram + density
plt.figure(figsize=(10, 6))

# Histogram normalized to density
plt.hist(speed, bins=20, edgecolor="black", alpha=0.5, density=True, label="Histogram")

# Density plot
speed.plot(kind="density", color="purple", linewidth=2, label="Density")

# Central tendency lines
plt.axvline(mean_speed, color="red", linestyle="--", label=f"Mean = {mean_speed:.2f}")
plt.axvline(median_speed, color="green", linestyle="--", label=f"Median = {median_speed:.2f}")
plt.axvline(mode_speed, color="blue", linestyle="--", label=f"Mode = {mode_speed:.2f}")

# Spread lines (±1 standard deviation from mean)
plt.axvline(mean_speed - std_speed, color="orange", linestyle=":", label=f"-1 Std = {(mean_speed - std_speed):.2f}")
plt.axvline(mean_speed + std_speed, color="orange", linestyle=":", label=f"+1 Std = {(mean_speed + std_speed):.2f}")

plt.title("Density Plot of Base Speed with Spread")
plt.xlabel("Speed")
plt.ylabel("Density")
plt.legend()
plt.show()

(mean_speed, median_speed, mode_speed, std_speed, var_speed, range_speed, iqr_speed, mad_speed)
```
