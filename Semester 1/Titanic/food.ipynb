What is the ratio of Google to Yelp ratings for the most popular restaurants in Denver 2025?(popular is anything above 4 stars.) This will be a retrospective study because the data already exists. Data was hand collected by me.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.read_csv("food - Sheet1.csv")
df=pd.read_csv("food - Sheet1.csv")

# Create scatterplot with regression line (line of best fit)
plt.figure(figsize=(8,6))
sns.regplot(x="Google Reviews", y="Yelp Reviews", data=df, scatter_kws={"s":70}, line_kws={"color":"red"})

plt.title("Google vs Yelp Reviews for Top Restaurants in Denver (2025)")
plt.xlabel("Google Reviews")
plt.ylabel("Yelp Reviews")
plt.grid(True)
plt.show()
```

The ratio of Google to Yelp reviews pretty even. There are some reviews that are higher or lower than others, but there is a pretty even spread.
